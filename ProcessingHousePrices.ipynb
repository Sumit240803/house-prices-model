```python
from scipy.stats import zscore
```

```python
z_score = zscore(data[columns])
outlier = (abs(z_score)>3)
data_outlier = data[outlier]
print(data_outlier)
```

```python
data_clean = data.dropna(subset=columns)  # Drop rows where any of the specified columns have NaN values
```

```python
data_clean
```

```python
z_scores = zscore(data['Price'])
outliers = data[abs(z_scores) > 3]
print(outliers.info())
```

```python
z_scores = zscore(data['lot area'])
outliers = data[abs(z_scores) > 3]
print(outliers.info())
```

```python
z_scores = zscore(data['number of bedrooms'])
outliers = data[abs(z_scores) > 3]
print(outliers.info())
```

```python
Q1 = data['Price'].quantile(0.25)
Q3 = data['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data_cleaned = data[(data['Price'] >= lower_bound) & (data['Price'] <= upper_bound)]
```

```python
data_cleaned.info()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Price'],kde=True, bins=90)
plt.title("Distribution of Price")
plt.show()
```
