```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
```

```python
X = pd.read_csv("../Datasets/HousePriceIndia/Cleaned-House-Prices.csv")
```

```python
Y =pd.read_csv("../Datasets/HousePriceIndia/ProcessedPrice.csv")
```

```python
Y = Y["Price"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X ,Y,test_size=0.2, random_state=42)
lr = LinearRegression()
lr.fit(X_train,Y_train)
y_pred_lr = lr.predict(X_test)
r2_score(Y_test,y_pred_lr)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, Y_train)
y_pred_rf = rf.predict(X_test)
r2_score(Y_test,y_pred_rf)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'max_depth': [15, 20, 25],
    'min_samples_leaf': [1, 2, 3],
    'max_features': ['sqrt', 0.8],
    'n_estimators': [100, 150, 200]
}


rf = RandomForestRegressor(random_state=42)
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_distributions,
    n_iter=20,  # Number of parameter combinations to try
    scoring='neg_mean_squared_error',  # Choose an appropriate scoring metric
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
random_search.fit(X_train,Y_train)
print("Best parameters found: ", random_search.best_params_)
```

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
print("Mean Squared Error:", mean_squared_error(Y_test, y_pred))
```

```python
from math import sqrt
```

```python
rmse = sqrt(7332571021.611472)
print("Root Mean Squared Error:", rmse)
```

```python
r2 = r2_score(Y_test, y_pred)
print("R² Score:", r2)
```

```python
train_score = best_model.score(X_train, Y_train)
test_score = best_model.score(X_test, Y_test)
print("Training R²:", train_score)
print("Test R²:", test_score)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(rf, X, Y, scoring='r2', cv=5)
print("Cross-validated R²:", cv_scores.mean())
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(max_depth=10,
    n_estimators=100,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42)
xgb.fit(X_train, Y_train)
```

```python
y_pred = xgb.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:", mse)

# R² Score
r2 = r2_score(Y_test, y_pred)
print("R² Score:", r2)
```

```python
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}
xgb_random_search = RandomizedSearchCV(xgb, param_dist, n_iter=50, cv=3, random_state=42)
xgb_random_search.fit(X_train, Y_train)
print("Best Parameters:", xgb_random_search.best_params_)
```

```python
xgb_best = XGBRegressor(
    subsample=0.9,
    n_estimators=200,
    max_depth=5,
    learning_rate=0.2,
    colsample_bytree=0.8,
    random_state=42
)

xgb_best.fit(X_train, Y_train)
```

```python
y_pred_best = xgb_best.predict(X_test)

# Calculate MSE and R²
mse_best = mean_squared_error(Y_test, y_pred_best)
r2_best = r2_score(Y_test, y_pred_best)

print(f"Best Model - MSE: {mse_best}")
print(f"Best Model - R²: {r2_best}")
```

```python
import matplotlib.pyplot as plt
from xgboost import plot_importance

# Plot importance using 'gain', 'weight', or 'cover'
plot_importance(xgb_best, importance_type='gain')  # Show gain-based importance
plt.show()

plot_importance(xgb_best, importance_type='weight')  # Show weight-based importance
plt.show()

plot_importance(xgb_best, importance_type='cover')  # Show cover-based importance
plt.show()

```
